```python
import PyPDF2

def load_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        text = ''
        for page in pdf_reader.pages:
            
            text += page.extract_text()
        return text


pdf_file = "c:/development/INPEX_Dataset_Analysis/data/DDR/160430_IOA_DMR08_BDC-1B-03_1795.pdf"

text_content = load_pdf(pdf_file)
print(text_content)
```

```python
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
import re

# Path to the PDF file
pdf_file_path = 'c:/development/INPEX_Dataset_Analysis/data/DDR/160430_IOA_DMR08_BDC-1B-03_1795.pdf'

# Convert the PDF to images
images = convert_from_path(pdf_file_path)

# Placeholder for extracted data
extracted_data = []

# Loop through each image (page)
for i, image in enumerate(images):
    # OCR process on the image
    text = pytesseract.image_to_string(image, lang='eng')

    # Add the text to our data array
    extracted_data.append(text)

# Since we're focusing on structured data extraction,
# we'll create regex patterns to match the data format we saw in the uploaded file
# and compile them into a dictionary.
structured_data = {
    'daily_time_analysis': [],
    'drilling_formation': [],
    'drilling': []
}

# Regex patterns to match the data blocks
pattern_daily_time_analysis = re.compile(
    r'DAILY TIME ANALYSIS \(Hours\): FORMATION\s+(.*?)\s+DRILLING FORMATION',
    re.DOTALL | re.IGNORECASE
)

pattern_drilling_formation = re.compile(
    r'DRILLING FORMATION\s+(.*?)\s+DRILLING',
    re.DOTALL | re.IGNORECASE
)

pattern_drilling = re.compile(
    r'DRILLING\s+(.*?)\s+DITCH MAGNET',
    re.DOTALL | re.IGNORECASE
)

# Function to extract data using regex patterns


def extract_data_by_pattern(text_block, pattern):
    matches = pattern.findall(text_block)
    return [match.strip() for match in matches] if matches else []


# Process each page of extracted text
for page_text in extracted_data:
    structured_data['daily_time_analysis'].extend(
        extract_data_by_pattern(page_text, pattern_daily_time_analysis))
    structured_data['drilling_formation'].extend(
        extract_data_by_pattern(page_text, pattern_drilling_formation))
    structured_data['drilling'].extend(
        extract_data_by_pattern(page_text, pattern_drilling))

# Let's see what we've got (just a preview, since the data could be quite extensive)
for key, data in structured_data.items():
    print(f"--- {key.upper()} ---")
    for entry in data:
        # print the first 100 characters to keep the output short
        print(entry[:100] + "...")
    print("\n")

# We could then parse these blocks further into more structured data such as CSV,
# but this would depend on the consistency and structure of the text within each block.
# That parsing would likely involve more complex regex to match each individual data field,
# which would require more information about the expected format.
```
